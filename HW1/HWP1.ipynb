3)Encode the target variable (y) .

4)Normalize the feature set (X).

5)Split the dataset into training and testing sets .

6)Implement and train a LinearRegression model, then use it to predict on the test data.

7)Implement and train a LogisticRegression model, then use it to predict on the test data.

8)Implement and train an SGDClassifier with logistic loss, then use it to predict on the test data.

9)Evaluate the modelsâ€™ performance by calculating accuracy or other metrics.

10)Save and compare the actual and predicted labels from all models in a DataFrame.

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression, SGDClassifier
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

1. Load the Iris Dataset

```python
iris = load_iris()
df = pd.DataFrame(data=iris.data , columns=iris.feature_names)
print(df.head())
print(iris.target)
```

2. Select Features and Target

```python
x = iris.data
y = iris.target
print(x)
print(y)
```

3. Encode the Target Variable

```python
y_names = iris.target_names
print(y_names)
y_all = [y_names[names] for names in y]
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y_all)
print(y_encoded)
```

4. Normalize Features

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(x)
print(X_scaled)
```

5. Split the Dataset

```python
x_train,x_test,y_train,y_test= train_test_split(X_scaled,y_encoded,train_size=0.8,random_state=42)
```

6. Implement Linear Regression

```python
linear_model = LinearRegression()
linear_model.fit(x_train,y_train)
y_pred_linear = linear_model.predict(x_test)
print(y_pred_linear)
```

7. Implement Logistic Regression

```python
logistic_model = LogisticRegression(max_iter=200)
logistic_model.fit(x_train,y_train)
y_pred_logistic = logistic_model.predict(x_test)
print(y_pred_logistic)
```

8. Implement SGD (Stochastic Gradient Descent)

```python
sgd= SGDClassifier(loss='log_loss', max_iter=200, tol=1e-3, random_state=42)
sgd.fit(x_train,y_train)
y_pred_sgd = sgd.predict(x_test)
print(y_pred_sgd)
```

9. Evaluate the Models

```python
from sklearn.metrics import accuracy_score
y_pred_linear_round = np.round(y_pred_linear)
acc_linear_regression= accuracy_score(y_test , y_pred_linear_round)
acc_logistic_regression=accuracy_score(y_test,y_pred_logistic)
acc_sgd=accuracy_score(y_test,y_pred_sgd)
print("Linear Regression Accuracy:",acc_linear_regression)
print("Logistic Regression Accuracy:",acc_logistic_regression)
print("SGD Accuracy:" ,acc_sgd)
```

```python
from sklearn.model_selection import cross_val_score
cv_scores_linear = cross_val_score(linear_model, X_scaled, y_encoded, cv=5)
cv_scores_logistic = cross_val_score(logistic_model, X_scaled, y_encoded, cv=5)
cv_scores_sgd = cross_val_score(sgd, X_scaled, y_encoded, cv=5)
print("Linear Regression Accuracy:", cv_scores_linear.mean())
print("Logistic Regression Accuracy:", cv_scores_logistic.mean())
print("SGD Accuracy:", cv_scores_sgd.mean())
```

10. Save Predictions

```python
y_pred_linear_round = np.round(y_pred_linear)
submission = pd.DataFrame({
    'Actual': y_test,
    'Predicted_Linear': y_pred_linear_round,
    'Predicted_Logistic': y_pred_logistic,
    'Predicted_SGD': y_pred_sgd
})
print(submission.head())
```
